```
import cv2
import numpy as np
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import imutils

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

```
url = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/text/text1.jpg"
img = url_to_image(url)
grayscaled = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
cv2_imshow(grayscaled)

retval, threshold = cv2.threshold(grayscaled, 160, 255, cv2.THRESH_BINARY)
cv2_imshow(threshold)

print("Result:")
denoised = cv2.fastNlMeansDenoising(threshold,None,53,5,17)
cv2_imshow(denoised)
```

```
url = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/text/text2.jpg"
img = url_to_image(url)
shifted = cv2.pyrMeanShiftFiltering(img, 27, 31)
grayscaled = cv2.cvtColor(shifted,cv2.COLOR_RGB2GRAY)
cv2_imshow(grayscaled)

blur = cv2.GaussianBlur(grayscaled, (11,11), 15)
# retval, threshold = cv2.threshold(grayscaled, 120, 255, cv2.THRESH_BINARY)
threshold = cv2.adaptiveThreshold(grayscaled, 255,
                          cv2.ADAPTIVE_THRESH_MEAN_C,
                          cv2.THRESH_BINARY, 13, 5)
# threshold = cv2.fastNlMeansDenoising(threshold,None,53,5,17)
# kernel = np.ones((5,5))
# denoised = cv2.erode(denoised, kernel,iterations=1)
# denoised = cv2.dilate(denoised, kernel,iterations=1)
cv2_imshow(threshold)

inv = cv2.bitwise_not(threshold)
cv2_imshow(inv)

ret, labels = cv2.connectedComponents(threshold, connectivity=8)

# Map component labels to hue val
label_hue = np.uint8(179*labels/np.max(labels))
blank_ch = 255*np.ones_like(label_hue)
labeled_img = cv2.merge([label_hue, blank_ch, blank_ch])

# cvt to BGR for display
labeled_img = cv2.cvtColor(labeled_img, cv2.COLOR_HSV2BGR)

# set bg label to black
labeled_img[label_hue==0] = 0
labeled_img[label_hue!=0] = 255

labels[labels>10] = 255

#cv2_imshow(img)
# cv2_imshow(labeled_img)

print("Result:")
cv2_imshow(labels)
```
