```
import numpy as np
import cv2
from skimage import io
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow

def url_to_image(url):
  print("downloading %s" % (url))
  return cv2.cvtColor(io.imread(url), cv2.COLOR_BGR2RGB)
```

```
url = "https://raw.githubusercontent.com/mdavydov/ComputerVisionCourse/master/tasks/object/obj1.jpg"
img = url_to_image(url)
img = cv2.resize(img, (0,0), fx=0.5, fy=0.5) 


(w,h) = img.shape[:2]
mask = np.zeros( (w,h),np.uint8)

bgdModel = np.zeros((1,65),np.float64)
fgdModel = np.zeros((1,65),np.float64)

x1 = 15
y1 = 5
hr = h - x1 - 15
wr = w - y1 - 10

print(img.shape)

rect = (x1,y1,hr,wr)
imgc = img.copy()
cv2.rectangle(imgc, (x1,y1), (hr+x1,wr+y1), (255,0,0), 2)
cv2_imshow(imgc)

cv2.grabCut(img,mask,rect,bgdModel,fgdModel,7,cv2.GC_INIT_WITH_RECT)
mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
img = img*mask2[:,:,np.newaxis]
cv2_imshow(mask2*255)

cv2_imshow(img)
```

### Universal solution

```
def separate(url, iter_num=5):
  img = url_to_image(url)
  img = cv2.resize(img, (0,0), fx=0.5, fy=0.5) 


  (w,h) = img.shape[:2]
  mask = np.zeros( (w,h),np.uint8)

  bgdModel = np.zeros((1,65),np.float64)
  fgdModel = np.zeros((1,65),np.float64)

#   x1 = 15
#   y1 = 5
#   hr = h - x1 - 10
#   wr = w - y1

  print((w,h))
  x1 = int(h * 0.05)
  y1 = int(w * 0.05)
  hr = h - x1 - int(h * 0.05)
  wr = w - y1

  print(img.shape)

  rect = (x1,y1,hr,wr)
  imgc = img.copy()
  cv2.rectangle(imgc, (x1,y1), (hr+x1,wr+y1), (255,0,0), 2)
  cv2_imshow(imgc)

  cv2.grabCut(img,mask,rect,bgdModel,fgdModel,iter_num,cv2.GC_INIT_WITH_RECT)
  mask2 = np.where((mask==2)|(mask==0),0,1).astype('uint8')
  img = img*mask2[:,:,np.newaxis]
  cv2_imshow(mask2*255)

  cv2_imshow(img)
```
